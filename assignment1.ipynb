```python
# Q1
import numpy as np
import math

# (a) Create a 1D NumPy array with values 1 to 20
arr = np.arange(1, 21)
print("Array from 1 to 20:")
print(arr)

# (b) Extract all prime numbers from it
isprime = []
for x in arr:
    if x < 2:
        continue
    for i in range(2, int(math.sqrt(x)) + 1):
        if x % i == 0:
            break
    else:
        isprime.append(x)

primes = np.array(isprime)    
print("\nPrime numbers:")
print(primes)

# (c) Compute the mean and variance of the extracted primes
mean_primes = np.mean(primes)
variance_primes = np.var(primes)
print(f"\nMean of primes: {mean_primes}")
print(f"Variance of primes: {variance_primes}")
```

```python
# Q4
# (a) Simulate 1000 coin tosses using NumPy (1=Head, 0=Tail)
coin_tosses = np.random.randint(0, 2, size=1000)
print("First 20 coin tosses:", coin_tosses[:20])

# (b) Count frequency of heads and tails
heads = np.sum(coin_tosses == 1)
tails = np.sum(coin_tosses == 0)
print(f"\nHeads: {heads}")
print(f"Tails: {tails}")

# (c) Estimate probability of heads. Is it close to 0.5? Why/Why not?
prob_heads = heads / 1000
print(f"\nProbability of heads: {prob_heads}")
print(f"Difference from 0.5: {abs(prob_heads - 0.5)}")
print("\nAnalysis: The probability is close to 0.5 due to the Law of Large Numbers.")
print("With 1000 tosses, the empirical probability converges to the theoretical probability (0.5).")
print("Small deviations are expected due to random variation.")
```

```python
# Q9
# (a) Generate 500 random integers between 1 and 6 (simulate dice rolls)
dice_rolls = np.random.randint(1, 7, size=500)
print("First 30 dice rolls:", dice_rolls[:30])

# (b) Count how many times each face appears
unique, counts = np.unique(dice_rolls, return_counts=True)
print("\nFrequency of each face:")
for face, count in zip(unique, counts):
    print(f"Face {face}: {count} times")

# (c) Compute relative frequencies and compare with theoretical 1/6
print("\nRelative frequencies compared to theoretical (1/6 = 0.1667):")
for face, count in zip(unique, counts):
    relative_freq = count / 500
    theoretical = 1/6
    difference = abs(relative_freq - theoretical)
    print(f"Face {face}: {relative_freq:.4f} (difference: {difference:.4f})")

print("\nAnalysis: With 500 rolls, each face should appear approximately 83-84 times.")
print("The relative frequencies should be close to 0.1667 (1/6) due to the Law of Large Numbers.")
```

### Case Study 1 – Student Performance Analytics

You are given data of 100 students containing their marks in **Math, Science, and English**.

```python
# Case Study 1 - Student Performance Analytics
import matplotlib.pyplot as plt

# Generate data for 100 students
np.random.seed(42)
num_students = 100

case1_data = {
    'Student_ID': range(1, num_students + 1),
    'Math': np.random.randint(40, 100, num_students),
    'Science': np.random.randint(35, 100, num_students),
    'English': np.random.randint(45, 100, num_students)
}
df_case1 = pd.DataFrame(case1_data)

print("First 10 students:")
print(df_case1.head(10))

# (a) Calculate average marks for each student and identify top 5 performers
df_case1['Average'] = df_case1[['Math', 'Science', 'English']].mean(axis=1)
top_5 = df_case1.nlargest(5, 'Average')
print("\nTop 5 performers:")
print(top_5[['Student_ID', 'Math', 'Science', 'English', 'Average']])

# (b) Compute subject-wise average and standard deviation; comment on variation
subject_stats = pd.DataFrame({
    'Subject': ['Math', 'Science', 'English'],
    'Average': [df_case1['Math'].mean(), df_case1['Science'].mean(), df_case1['English'].mean()],
    'Std_Deviation': [df_case1['Math'].std(), df_case1['Science'].std(), df_case1['English'].std()]
})
print("\nSubject-wise statistics:")
print(subject_stats)

max_std = subject_stats.loc[subject_stats['Std_Deviation'].idxmax(), 'Subject']
print(f"\n{max_std} has the highest variation, indicating wider spread in student performance.")

# (c) Draw a boxplot for each subject and comment on outliers
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
subjects = ['Math', 'Science', 'English']

for i, subject in enumerate(subjects):
    axes[i].boxplot(df_case1[subject])
    axes[i].set_title(f'{subject} Marks Distribution')
    axes[i].set_ylabel('Marks')
    axes[i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nOutlier Analysis:")
for subject in subjects:
    Q1 = df_case1[subject].quantile(0.25)
    Q3 = df_case1[subject].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df_case1[(df_case1[subject] < Q1 - 1.5*IQR) | (df_case1[subject] > Q3 + 1.5*IQR)]
    print(f"{subject}: {len(outliers)} outliers detected")
```

### Case Study 2 – Sales Data Exploration

A company records sales transactions in a dataset with columns: **Product, Quantity, Price, Region**.

```python
# Case Study 2 - Sales Data Exploration

# Create sample sales dataset
np.random.seed(42)
products = ['Laptop', 'Phone', 'Tablet', 'Headphones', 'Keyboard', 'Mouse']
regions = ['North', 'South', 'East', 'West']

case2_data = {
    'Product': np.random.choice(products, 100),
    'Quantity': np.random.randint(1, 50, 100),
    'Price': np.random.randint(500, 50000, 100),
    'Region': np.random.choice(regions, 100)
}
df_case2 = pd.DataFrame(case2_data)

print("First 10 sales transactions:")
print(df_case2.head(10))

# (a) Add a new column Total = Quantity × Price and compute overall revenue
df_case2['Total'] = df_case2['Quantity'] * df_case2['Price']
overall_revenue = df_case2['Total'].sum()
print(f"\nOverall Revenue: ₹{overall_revenue:,}")

# (b) Group data by region and identify which region contributes the most sales
region_sales = df_case2.groupby('Region')['Total'].sum().sort_values(ascending=False)
print("\nSales by Region:")
print(region_sales)
print(f"\nTop contributing region: {region_sales.idxmax()} with ₹{region_sales.max():,}")

# (c) Plot a histogram of total sales per product and discuss underperforming products
product_sales = df_case2.groupby('Product')['Total'].sum().sort_values(ascending=False)
print("\nSales by Product:")
print(product_sales)

plt.figure(figsize=(12, 6))
product_sales.plot(kind='bar', color='steelblue')
plt.title('Total Sales Revenue by Product', fontsize=14, fontweight='bold')
plt.xlabel('Product')
plt.ylabel('Total Revenue (₹)')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

# Identify underperforming products
avg_sales = product_sales.mean()
underperforming = product_sales[product_sales < avg_sales]
print(f"\nAverage product sales: ₹{avg_sales:,.2f}")
print("\nUnderperforming products (below average):")
for product, sales in underperforming.items():
    print(f"  {product}: ₹{sales:,} ({(sales/avg_sales - 1)*100:.1f}% below average)")
```

### Case Study 3 – Predicting Fuel Efficiency (Auto MPG Dataset)

You are analyzing the **Auto MPG dataset (UCI Repository)** with features: mpg, cylinders, displacement, horsepower, weight, acceleration, model year, origin.

```python
# Case Study 3 - Predicting Fuel Efficiency (Auto MPG Dataset)

# Load the Auto MPG dataset
try:
    df_mpg = pd.read_csv('auto-mpg.csv')
    print("Auto MPG Dataset loaded successfully!")
    print(f"Dataset shape: {df_mpg.shape}")
    print("\nFirst few rows:")
    print(df_mpg.head())
    print("\nDataset info:")
    print(df_mpg.info())
except FileNotFoundError:
    # If auto-mpg.csv doesn't exist, download it or create sample data
    print("Loading Auto MPG dataset from UCI repository...")
    url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'
    column_names = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 
                    'acceleration', 'model_year', 'origin', 'car_name']
    df_mpg = pd.read_csv(url, names=column_names, delim_whitespace=True, na_values='?')
    df_mpg.to_csv('auto-mpg.csv', index=False)
    print("Dataset downloaded and saved!")
    print(f"Dataset shape: {df_mpg.shape}")
    print("\nFirst few rows:")
    print(df_mpg.head())
```

```python
# (a) Compute correlation of mpg with each numeric feature and discuss which factors influence fuel efficiency most

# Select numeric columns only
numeric_cols = df_mpg.select_dtypes(include=[np.number]).columns.tolist()
print("Numeric columns:", numeric_cols)

# Compute correlation with mpg
if 'mpg' in numeric_cols:
    correlations = df_mpg[numeric_cols].corr()['mpg'].sort_values(ascending=False)
    print("\nCorrelation of MPG with other features:")
    print(correlations)
    
    print("\nInterpretation:")
    print("=" * 60)
    for feature, corr in correlations.items():
        if feature != 'mpg':
            if abs(corr) > 0.7:
                strength = "Strong"
            elif abs(corr) > 0.4:
                strength = "Moderate"
            else:
                strength = "Weak"
            
            direction = "positive" if corr > 0 else "negative"
            print(f"{feature:15s}: {corr:6.3f} - {strength} {direction} correlation")
    
    # Identify most influential factors
    top_factors = correlations[correlations.index != 'mpg'].abs().sort_values(ascending=False).head(3)
    print("\nMost influential factors for fuel efficiency:")
    for factor, corr in top_factors.items():
        print(f"  {factor}: {correlations[factor]:.3f}")
else:
    print("MPG column not found in dataset")
```

```python
# (b) Plot a scatter plot of weight vs mpg; describe the trend you observe

plt.figure(figsize=(10, 6))

# Handle potential missing values
if 'weight' in df_mpg.columns and 'mpg' in df_mpg.columns:
    # Remove rows with missing values for this plot
    plot_data = df_mpg[['weight', 'mpg']].dropna()
    
    plt.scatter(plot_data['weight'], plot_data['mpg'], alpha=0.6, c='darkblue', edgecolors='black')
    plt.title('Relationship between Vehicle Weight and Fuel Efficiency', fontsize=14, fontweight='bold')
    plt.xlabel('Weight (lbs)', fontsize=12)
    plt.ylabel('MPG (Miles Per Gallon)', fontsize=12)
    plt.grid(True, alpha=0.3)
    
    # Add trend line
    z = np.polyfit(plot_data['weight'], plot_data['mpg'], 1)
    p = np.poly1d(z)
    plt.plot(plot_data['weight'], p(plot_data['weight']), "r--", linewidth=2, label=f'Trend line: y={z[0]:.4f}x+{z[1]:.2f}')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
    # Calculate correlation
    corr = plot_data['weight'].corr(plot_data['mpg'])
    print(f"\nCorrelation between weight and mpg: {corr:.4f}")
    
    print("\nTrend Analysis:")
    print("=" * 60)
    print("There is a strong NEGATIVE correlation between weight and MPG.")
    print("As vehicle weight increases, fuel efficiency (MPG) decreases.")
    print("This makes intuitive sense: heavier vehicles require more energy")
    print("to move, resulting in lower miles per gallon.")
    print(f"The trend line shows approximately {abs(z[0]):.4f} MPG decrease")
    print("for every 1 lb increase in vehicle weight.")
else:
    print("Weight or MPG column not found in dataset")
```
